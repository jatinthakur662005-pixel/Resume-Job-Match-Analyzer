```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    return text

resume_path = 'Jatin Kumar Singh Resume.pdf'
resume_text = extract_text_from_pdf(resume_path)

print("Resume text extracted successfully!")
print(resume_text[:500])  # preview first 500 characters
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded successfully!")
```

```python
job_description = """
We are looking for a Data Analyst with strong skills in SQL, Python,
Excel, and Power BI. The candidate should have experience in data
cleaning, statistical analysis, and creating dashboards. Strong
communication skills and ability to work with cross-functional teams
required. Knowledge of machine learning is a plus.
"""
print("Job description loaded.")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

resume_embedding = model.encode([resume_text])
jd_embedding = model.encode([job_description])

similarity_score = cosine_similarity(resume_embedding, jd_embedding)[0][0]
print(f"Match Score: {similarity_score * 100:.2f}%")
```

```python
def extract_keywords(text):
    common_skills = [
        'python', 'sql', 'excel', 'power bi', 'tableau', 'machine learning',
        'communication', 'data analysis', 'statistics', 'r programming',
        'aws', 'deep learning', 'nlp', 'data visualization', 'analytics',
        'java', 'javascript', 'react', 'node.js', 'git', 'docker',
        'leadership', 'teamwork', 'project management'
    ]
    text_lower = text.lower()
    found = set(skill for skill in common_skills if skill in text_lower)
    return found

resume_skills = extract_keywords(resume_text)
jd_skills = extract_keywords(job_description)

matching_skills = resume_skills & jd_skills
missing_skills = jd_skills - resume_skills

print("✅ Matching Skills:", matching_skills if matching_skills else "None found")
print("❌ Missing Skills:", missing_skills if missing_skills else "None")
```

```python
def analyze_resume_match(resume_text, job_description):
    resume_emb = model.encode([resume_text])
    jd_emb = model.encode([job_description])
    score = cosine_similarity(resume_emb, jd_emb)[0][0] * 100

    resume_skills = extract_keywords(resume_text)
    jd_skills = extract_keywords(job_description)
    matching = resume_skills & jd_skills
    missing = jd_skills - resume_skills

    print(f"Match Score: {score:.2f}%")
    print("✅ Matching Skills:", matching if matching else "None")
    print("❌ Missing Skills:", missing if missing else "None")
    return score, matching, missing

# Test it
analyze_resume_match(resume_text, job_description)
```

```python
job_description_2 = """
Seeking a Business Analyst with expertise in stakeholder management,
requirements gathering, SQL, Excel, and data-driven decision making.
Experience with Agile methodology and Jira preferred.
"""

analyze_resume_match(resume_text, job_description_2)
```
